```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os

# File to Load (Remember to Change These)
school_data_to_load = os.path.join('Resources', 'schools_complete.csv')
student_data_to_load = os.path.join('Resources', 'students_complete.csv')

# Read School and Student Data File and store into Pandas DataFrames
school_df = pd.read_csv(school_data_to_load)
student_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
complete_df = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
###--------- Start of District Summary ---------

#Calculate total # of schools in the complete data, use the .nunique function:
total_schools = complete_df['school_name'].nunique()

#Calc total # of students in the complete data, use the .count function:
total_students = complete_df['student_name'].count()

#Calc total budget from the school data, use .sum function:
total_budget = school_df["budget"].sum()

#Calc the average math score, use the .mean function:
avg_math = complete_df['math_score'].mean()

#Calc the average reading score, use the .mean function:
avg_reading = complete_df['reading_score'].mean()

#Find the students w/ passing math score (70 or greater):
passing_math_scores = complete_df.loc[complete_df['math_score'] >= 70]

#Count of the total math scores:
total_math_scores = complete_df['math_score'].count()

#Count of the passing math scores:
total_passing_math = passing_math_scores['math_score'].count()

#Calc the % of students w/ passing math score (70 or greater):
pct_passing_math = (total_passing_math / total_math_scores)

#Find the students w/ passing reading score (70 or greater):
passing_read_scores = complete_df.loc[complete_df['reading_score'] >= 70]
#Count of the total reading scores:
total_read_scores = complete_df['reading_score'].count()

#Count of the passing reading scores:
total_passing_read = passing_read_scores['reading_score'].count()

#Calc the % of students w/ passing reading score (70 or greater):
pct_passing_read = (total_passing_read / total_read_scores)

#Find the students who pass math AND reading:
overall_passing = passing_math_scores.loc[(passing_math_scores['math_score'] >= 70) & 
                                                (passing_read_scores['reading_score'] >= 70)]

#Find the total of students who pass math:
total_passing_math = overall_passing['math_score'].count()

#Find the total of students who pass reading:
total_passing_read = overall_passing['reading_score'].count()

#Calc total of students who pass math AND reading:
total_overall_passing = total_passing_math + total_passing_read

#Calc total math AND read scores (not just passing):
total_scores = total_math_scores + total_read_scores

#Finding the overall % passing:
pct_overall_pass = (total_overall_passing / total_scores)

#Create dataframe to display 'DISTRICT SUMMARY':
district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math],
    "Average Reading Score": [avg_reading],
    "% Passing Math": [pct_passing_math],
    "% Passing Reading": [pct_passing_read],
    "% Overall Passing": [pct_overall_pass]
})

district_summary.style.format({
    "Total Students": "{:,}",
    "Total Budget": "${:,.2f}",
    "Average Math Score": "{:.3f}",
    "Average Reading Score": "{:.3f}",
    "% Passing Math": "{:.3%}",
    "% Passing Reading": "{:.3%}",
    "% Overall Passing": "{:.3%}"
})

###--------- End of District Summary ---------
```

```python
###--------- Start of School Summary ---------
                           
#Grouped 'school_name' and 'type' to be able to find the calculations for the summary:
grouped_df = complete_df.groupby(['school_name'])


# >>> 'School Type' <<<
sch_type = grouped_df['type'].first()

# >>> 'Total Students' <<<
tot_stu = grouped_df['size'].count()

# >>> 'Total School Budget' <<<
tot_sch_budget = grouped_df['budget'].mean()

# >>> 'Per Student Budget' <<<
per_student_budget = (tot_sch_budget / tot_stu)

#Now we can find the scores per school based on student's name:
tot_scores = grouped_df['student_name'].count()

# >>> 'Average Math Score' <<<
avg_math_score = grouped_df['math_score'].mean()
# >>> 'Average Reading Score' <<<
avg_read_score = grouped_df['reading_score'].mean()

#Grouping the math, reading, and overall passing scores, respectively by 'school_name':
group_math_df = passing_math_scores.groupby('school_name')
group_read_df = passing_read_scores.groupby('school_name')
group_scores_df = overall_passing.groupby('school_name')

num_of_passing_math = group_math_df.count()['student_name']
num_of_passing_read = group_read_df.count()['student_name']
num_of_overall_passing = group_scores_df.count()['student_name']

#Calc - % for each category (math, reading, and overall):
# >>> '% Passing Math' <<<
pct_pass_math = (num_of_passing_math / tot_scores) * 100
# >>> '% Passing Reading' <<<
pct_pass_read = (num_of_passing_read / tot_scores) * 100
# >>> '% Overall Passing' <<<
pct_overall_pass = (num_of_overall_passing / tot_scores) * 100

#Create dataframe to display 'SCHOOL SUMMARY':
school_summary = pd.DataFrame({
    "School Type": sch_type,
    "Total Students": tot_stu,
    "Total Budget": tot_sch_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": avg_math_score,
    "Average Reading Score": avg_read_score,
    "% Passing Math": pct_pass_math,
    "% Passing Reading": pct_pass_read,
    "% Overall Passing": pct_overall_pass
})
school_summary.style.format({
    "Total Students": "{:,}",
    "Total Budget": "${:,.2f}",
    "Per Student Budget": "${:,.2f}",
    "Average Math Score": "{:.3f}",
    "Average Reading Score": "{:.3f}",
    "% Passing Math": "{:.6}",
    "% Passing Reading": "{:.6}",
    "% Overall Passing": "{:.6}"
})
#--------- End of School Summary ---------
```

## Top Performing Schools (By % Overall Passing)

```python
top_performing_df = school_summary.sort_values('% Overall Passing', ascending=False)
top_performing_df.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
grade_9th_df = complete_df.loc[complete_df['grade'] == '9th']
avg_9th_math_df = grade_9th_df.groupby(['school_name']).mean()['math_score']

grade_10th_df = complete_df.loc[complete_df['grade'] == '10th']
avg_10th_math_df = grade_10th_df.groupby(['school_name']).mean()['math_score']

grade_11th_df = complete_df.loc[complete_df['grade'] == '11th']
avg_11th_math_df = grade_11th_df.groupby(['school_name']).mean()['math_score']

grade_12th_df = complete_df.loc[complete_df['grade'] == '12th']
avg_12th_math_df = grade_12th_df.groupby(['school_name']).mean()['math_score']

math_scores_by_grade = pd.DataFrame({
    "9th Grade": avg_9th_math_df,
    "10th Grade": avg_10th_math_df,
    "11th Grade": avg_11th_math_df,
    "12th Grade": avg_12th_math_df
})
math_scores_by_grade.style.format({
    "9th Grade": "{:.6}",
    "10th Grade": "{:.6}",
    "11th Grade": "{:.6}",
    "12th Grade": "{:.6}"
})
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
grade_9th_df = complete_df.loc[complete_df['grade'] == '9th']
avg_9th_read_df = grade_9th_df.groupby(['school_name']).mean()['reading_score']

grade_10th_df = complete_df.loc[complete_df['grade'] == '10th']
avg_10th_read_df = grade_10th_df.groupby(['school_name']).mean()['reading_score']

grade_11th_df = complete_df.loc[complete_df['grade'] == '11th']
avg_11th_read_df = grade_11th_df.groupby(['school_name']).mean()['reading_score']

grade_12th_df = complete_df.loc[complete_df['grade'] == '12th']
avg_12th_read_df = grade_12th_df.groupby(['school_name']).mean()['reading_score']

read_scores_by_grade = pd.DataFrame({
    "9th Grade": avg_9th_read_df,
    "10th Grade": avg_10th_read_df,
    "11th Grade": avg_11th_read_df,
    "12th Grade": avg_12th_read_df
})
read_scores_by_grade.style.format({
    "9th Grade": "{:.6}",
    "10th Grade": "{:.6}",
    "11th Grade": "{:.6}",
    "12th Grade": "{:.6}"
})
```

## Scores by School Spending

# Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Creating bins to hold our data:
bins = [0, 584, 629, 644, 675]

#Naming (labeling) the bins to reference later:
labels = ["<$584", "$585-629", "$630-644", "$645-675"]

#Cut the data needed and insert in bins:
school_summary["Spending Ranges (Per Student)"] = pd.cut(school_summary["Per Student Budget"], bins, labels = labels)

#Create summary (pd.DataFrame) to print:
scores_by_sch_spending = school_summary
scores_by_sch_spending = scores_by_sch_spending.groupby("Spending Ranges (Per Student)")
scores_by_sch_spending[[
    "Average Math Score", 
    "Average Reading Score", 
    "% Passing Math", 
    "% Passing Reading", 
    "% Overall Passing"]].mean()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#Creating bins to hold our data:
bins = [0, 1000, 2000, 5000]

#Naming (labeling) the bins to reference later:
labels = ["Small (< 1000)", "Medium (1000 - 2000)", "Large (2000 - 5000)"]

#Cut the data needed and insert in bins:
school_summary["School Size"] = pd.cut(school_summary["Total Students"], bins, labels = labels)

#Create summary (pd.DataFrame) to print:
scores_by_sch_size = school_summary
scores_by_sch_size = scores_by_sch_size.groupby("School Size")
scores_by_sch_size[[
    "Average Math Score", 
    "Average Reading Score", 
    "% Passing Math", 
    "% Passing Reading", 
    "% Overall Passing"]].mean()
```
